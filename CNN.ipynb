```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
tf.__version__
```

```python
training_datagen = ImageDataGenerator(
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)

training_set = training_datagen.flow_from_directory(
    'dataset/training_set',
    target_size =(64,64),
    batch_size =32 ,
    class_mode = 'binary'
)
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
    'dataset/test_set',
    target_size = (64,64),
    batch_size = 32,
    class_mode ='binary')
```

```python
cnn = tf.keras.models.Sequential()
```

```python
cnn.add(tf.keras.layers.Conv2D(filters =32 , activation='relu' , kernel_size= 3 , input_shape = [64,64,3] ))
```

```python
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2))
```

```python
cnn.add(tf.keras.layers.Conv2D(filters =32 , activation='relu' , kernel_size= 3 ))
cnn.add(tf.keras.layers.MaxPooling2D(pool_size=(2, 2), strides=2))
```

```python
cnn.add(tf.keras.layers.Flatten())
```

```python
cnn.add(tf.keras.layers.Dense(units=128 , activation='relu' , ))
```

```python
cnn.add(tf.keras.layers.Dense(units=1 , activation='sigmoid' , ))
```

```python
cnn.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics =['accuracy'])
```

```python
cnn.fit(x =training_set ,validation_data = test_set , epoch =25)
```
