```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv("/content/Ads_CTR_Optimisation.csv")
```

```python
import math
N = 10000
d = 10
ads_selected = []
number_of_selection = [0] * d
sum_of_rewards = [0] * d
total_rewards = 0
for n in range(N):
  ad = 0
  max_upper_bound = 0
  for i in range(d):
    if (number_of_selection[i] > 0):
      average_reward = sum_of_rewards[i] / number_of_selection[i]
      delta_i = math.sqrt((3/2) * math.log(n + 1) / number_of_selection[i])
      UBC = average_reward + delta_i
    else:
      UBC = 1e400
    if UBC > max_upper_bound:
      max_upper_bound = UBC
      ad = i
  ads_selected.append(ad)
  number_of_selection[ad] += 1
  reward = dataset.values[n,ad]
  sum_of_rewards[ad] += reward
  total_rewards += reward

plt.hist(ads_selected)
plt.title('Histogram of ads selections')
plt.xlabel('Ads')
plt.ylabel('Number of times each ad was selected')
plt.show()
```
